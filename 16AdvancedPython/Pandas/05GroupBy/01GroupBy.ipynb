Pandas Group By
---------------
Weather data from various cities and see how group by can be used to run some analytics.

```python
import pandas as pd
```

```python
df = pd.read_csv("weather_by_cities.csv")

df
```

```python
g = df.groupby("city")

for i in g:
    print(i)
```

```python
for city, data in g:
    print("city:",city)
    print("\n")
    print("data:")
    print(data)  
    print('')
```

```python
# Create group 

g.get_group('mumbai')
```
