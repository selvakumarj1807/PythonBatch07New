```python
import pandas as pd
```

```python
df = pd.read_csv('nyc_weather.csv')

df
```

To find Max temperature value in month of January:
--------------------------------------------------

```python
max_temp = df['Temperature'].max()

print('Max temperature in New York in month of January : ', max_temp)
```

```python
min_temp = df['Temperature'].min()

print('Min temperature in New York in month of January : ', min_temp)
```

```python
avg_temp = df['Temperature'].mean()

print('Min temperature in New York in month of January : ', avg_temp)
```

List of days when it rained:
----------------------------

```python
RainDates = df['EST'][df['Events'] == 'Rain']

RainDates
```
